```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Dados fornecidos pelo HRBP
data = {
    'Mês': ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun'],
    'Atrito': [18, 10, 13, 15, 22, 17],
    'Contratações': [25, 20, 17, 23, 19, 15],
    'Investimento em Treinamento': [3, 4, 2, 4, 5, 3]
}

# Criando um DataFrame com os dados
df = pd.DataFrame(data)

# Configuração dos dados
labels = df['Mês']
atrito = df['Atrito']
contratacoes = df['Contratações']
investimento = df['Investimento em Treinamento']

# Configuração da largura das barras
bar_width = 0.3

# Configuração dos índices das barras
index = np.arange(len(labels))

# Plotando o gráfico de barras em barras
plt.figure(figsize=(10, 6))

plt.bar(index, atrito, width=bar_width, label='Atrito', color='blue')
plt.bar(index + bar_width, contratacoes, width=bar_width, label='Contratações', color='green')
plt.bar(index + 2 * bar_width, investimento, width=bar_width, label='Investimento em Treinamento', color='red')

# Configurações do gráfico
plt.xlabel('Mês')
plt.ylabel('Quantidade')
plt.title('Atrito, Contratações e Investimento em Treinamento')
plt.xticks(index + bar_width, labels)
plt.legend()

# Exibindo o gráfico
plt.show()
```
